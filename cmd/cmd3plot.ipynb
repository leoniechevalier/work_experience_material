# This notebook is an example of how to plot a CMD


### Import numpy and pylab

```python
# This you need to do maths
import numpy as np
# This you need for plots
import matplotlib
matplotlib.use('TkAgg')
import pylab as plt
```

### Read in the cmd3.txt file with the data in it

```python
data = np.loadtxt('cmd3.txt')
```

### Print what is in the file

```python
print(data)
```

### make the CMD

```python
plt.figure()
plt.ylim(16.0,-2.0)
plt.scatter(data[:,2],data[:,1],label='star',color='green')
plt.xlabel('B-V',fontsize=12)
plt.ylabel('Mv',fontsize=12)
plt.legend()
plt.show()
```

### Add the distance modulus and reddening

```python
def AddDistanceModulus(a):
    dmv = 9.7
    c = a + dmv
    return c 
```

```python
def AddReddenning(b):
    ebminv = 0.05
    d = b + ebminv
    return d
```

```python
new_Mv = AddDistanceModulus(data[:,1]) 
```

```python
new_BminV = AddReddenning(data[:,2])
```

```python
plt.figure()
plt.ylim(25.0,5.0)
plt.scatter(new_BminV,new_Mv,label='star',color='blue')
plt.xlabel('B-V',fontsize=12)
plt.ylabel('Mv',fontsize=12)
plt.legend()
plt.show()
```

### Now plot the CMD with the model and observed data

```python
obsdata = np.loadtxt('m67temp.txt')
```

```python
plt.figure()
plt.ylim(25.0,5.0)
plt.scatter(obsdata[:,3],obsdata[:,2],label='cluster',color='red')
plt.scatter(new_BminV,new_Mv,label='model',color='blue')
plt.xlabel('B-V',fontsize=12)
plt.ylabel('Mv',fontsize=12)
plt.legend()
plt.show()
```
